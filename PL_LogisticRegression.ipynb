```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
bike_sharing = fetch_ucirepo(id=275)

# data (as pandas dataframes)
X = bike_sharing.data.features
y = bike_sharing.data.targets

# metadata
print(bike_sharing.metadata)

# variable information
print(bike_sharing.variables)
```

```python
# Drop non-numeric columns (e.g., date columns)
X = X.select_dtypes(include=[float, int])  # Keep only numeric columns

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 4: Preprocess the data (Normalize features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 5: Train the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_scaled, y)
```

```python
# Step 6: Evaluate the model
accuracy = model.score(X_scaled, y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Step 7: Predict and visualize (if necessary)
y_pred = model.predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap='viridis')
plt.title('Prediction Visualization')
plt.show()
```
